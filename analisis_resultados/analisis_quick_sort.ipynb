```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
tamanios = [1,1000,2500,5000,7500,10000,12500,15000,17500,20000]
yCython = []
yPython = []
cythonStd = []
pythonStd = []
```

```python
path = '/home/miguel/Desktop/computacion_paralela/final/quick_sort/soluciones'
paths = []
files = os.listdir(path)
for file in files:
    paths.append(str(path + '/' + file))
```

```python
paths.sort()
paths
```

```python
for path in paths:
    df = pd.read_csv(path)
    cythonStd.append(df['Cython'].std())
    pythonStd.append(df['Python'].std())
    yCython.append(df['Cython'].mean())
    yPython.append(df['Python'].mean())
```

```python
plt.plot(tamanios, yPython, label='Python')
plt.plot(tamanios, yCython, label='Cython')
plt.xlabel('Tamaño del arreglo (N)')
plt.ylabel('tiempo (s)')
plt.title("Python vs. Cython: Algoritmo Quick Sort")
plt.legend()
```

```python
res = []
for i in range(len(yPython)):
    veces = yPython[i] / yCython[i]
    res.append(veces)
    print(f'Cython es {veces} veces mas rapido que Python')

resultado = sum(res)/len(yCython)
print(resultado)
```

```python
fig, ax = plt.subplots()
x = np.arange(len(tamanios))
y1 = yCython
y2 = yPython
width = 0.2
plt.figure(figsize=(15,10))
# plot data in grouped manner of bar type
plt.bar(x, y1, width)
plt.bar(x-0.2, y2, width)
plt.xticks(x, tamanios)
plt.title("Cython vs. Python: Quick Sort Performance")
plt.xlabel("Tamaño del Arreglo (N)")
plt.ylabel("Segundos (s)")
plt.legend(['Cython', 'Python'])

```
